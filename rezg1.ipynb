```python
from sympy import symbols, expand, simplify, diff, integrate, latex, sin, cos, exp, lambdify
import math
from math import radians, sqrt, atan, degrees
from IPython.display import display, Math
import numpy as np
import matplotlib.pyplot as plt

d, ro, m, e, l_1, l_2, l_3, l_4, k_1, k_2, k_t, c_1, F_0, M_0, r_0, omega, pi, g, t, omega_d, zeta, omega_n, C_1, C_2, phi, theta = symbols('d ro m e l_1 l_2 l_3 l_4 k_1 k_2 k_t c_1 F_0 M_0 r_0 omega pi g t omega_d zeta omega_n C_1 C_2 phi theta')

expr = exp(-zeta*omega_n*t) * (C_1 * sin(omega_d*t) + C_2 * cos(omega_d*t)) + phi * sin(omega*t-theta)
display(Math(latex(expr)))
```

```python
ertekek ={
    zeta: 0.136174,
    omega_d: 10.772867,
    omega_n: 10.874161,
    C_1: -0.00042,
    C_2: -0.04379,
    phi: 0.003396,
    theta: 3.017675,
    omega: 28
    
}

fv = expr.subs(ertekek)

display(Math(latex(fv)))
print(latex(fv))
```

```python

f = lambdify(t, fv, 'numpy')

t_vals = np.linspace(0,10,1000)
y_vals = f(t_vals)

plt.plot(t_vals, y_vals)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid(True)
plt.savefig('figure.jpg', format='jpg')
plt.show()
```
